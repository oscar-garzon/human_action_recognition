# Reconocimiento de acciones humanas en imágenes

```python
import numpy as np
from PIL import Image
from os import listdir
import csv
```

Para este proyecto se utiliza el dataset __[Human Action Recognition (HAR)](https://www.kaggle.com/datasets/meetnagadia/human-action-recognition-har-dataset)__. 

Este datase contiene 15 clases diferentes de activades humanas. El dataset original tiene 12k+ imágenes incluyendo las de validación. Solo usamos 3000 por cuestiones de espacio en RAM. Cada imagen tiene solo una categoria de actividad humana.    
    

```python
def load_images(path):
    """Devuelve las imágenes con los pixeles ya normalizados y con un resize de (128, 64)"""
    files = listdir(path)
    imagenes = []

    for i,img in enumerate(files):
        if i == 3000:
            break
        img_path = path + img
        with Image.open(img_path) as imagen:
            img_rs = imagen.resize((128,64))
        imagenes.append(np.asarray(img_rs) / 255 )
        print(i, '/3000')

    return np.asarray(imagenes)    
```

```python
def load_labels(path):
    """Regresa las labels correspondientes a cada imagen"""
    labels = []
    with open(path) as file:
        csvFile = csv.reader(file)
        for i, img in enumerate(csvFile):
            if i == 3001:                  #Se usa 3001 porque la primera linea son los títulos
                break
            labels.append(img[1])
    return labels[1:]
```

```python
x_ds = load_images('Human Action Recognition/train/')
```

```python
y_ds = load_labels('Human Action Recognition/Training_set.csv')
```

```python
def numero_asociado_a_label(label_str):
    """ Regresa el numero asociado a la categoria de la etiqueta"""
    labels =  { 
        'calling': 0.0, 
        'clapping': 1.0,
        'cycling': 2.0,
        'dancing': 3.0,
        'drinking': 4.0, 
        'eating': 5.0,
        'fighting': 6.0,
        'hugging': 7.0,
        'laughing': 8.0,
        'listening_to_music': 9.0,
        'running': 10.0,
        'sitting': 11.0,
        'sleeping': 12.0,
        'texting': 13.0, 
        'using_laptop': 14.0
    }
    
    return labels[label_str]
```

### Arquitectura

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras import layers
from tensorflow.keras.layers import Input, Dense, Activation, Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.imagenet_utils import preprocess_input
import pydot
from IPython.display import SVG
from tensorflow.keras.utils import plot_model
from tensorflow.keras.optimizers import SGD

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization

from tensorboard.plugins.hparams import api as hp

import time
```

Veamos que hiperparámetros funcionan mejor con el modelo

```python
HP_NUM_UNITS = hp.HParam('num_units', hp.Discrete([16, 32]))
HP_DROPOUT = hp.HParam('dropout', hp.Discrete([0.1, 0.4]))
HP_OPTIMIZER = hp.HParam('optimizer', hp.Discrete(['adam', 'sgd']))
HP_LEARNINGRATE= hp.HParam('learning_rate', hp.Discrete([0.1,0.01]))

METRIC_ACCURACY = 'accuracy'

with tf.summary.create_file_writer('logs/hparam_tuning').as_default():
  hp.hparams_config(
    hparams=[HP_NUM_UNITS, HP_DROPOUT, HP_OPTIMIZER, HP_LEARNINGRATE],
    metrics=[hp.Metric(METRIC_ACCURACY, display_name='Accuracy')],
  )
```

```python
def architecture(input_shape, num_classes, units, 
                 activation, dropout=True, 
                 batch_normalization=True,
                 pdrop=0):
    
    model = Sequential()
    
    #First convolution module    
    model.add(Conv2D(units, 3, padding='same', input_shape=input_shape))
    model.add(Activation(activation))
    model.add(MaxPooling2D())
    if batch_normalization: 
        model.add(BatchNormalization())  # Performing batch normalization to the output of the input layer
       
    #Second convolution module
    model.add(Conv2D(units, 3, padding='same',kernel_initializer='he_uniform',bias_initializer='zeros'))
    model.add(Activation(activation))
    model.add(MaxPooling2D())
    if batch_normalization:
        model.add(BatchNormalization())  # Performing batch normalization to the convolution layer
            
    #Third convolution module 
    model.add(Conv2D(2*units, 3, padding = 'same', kernel_initializer = 'he_uniform',bias_initializer='zeros'))
    model.add(Activation(activation))
    if batch_normalization:
        model.add(BatchNormalization())  # Performing batch normalization to the convolution layer
        
    #Fourth convolution module
    model.add(Conv2D(2*units, 3, padding = 'same', kernel_initializer = 'he_uniform',bias_initializer='zeros'))
    model.add(Activation(activation))
    model.add(MaxPooling2D())
    if batch_normalization:
        model.add(BatchNormalization())  # Performing batch normalization to the convolution layer
    

    #Feed forward "full" connected module
    model.add(Flatten())
    if dropout:
        model.add(Dropout(pdrop))
        
    model.add(Dense(50, kernel_initializer='he_uniform',bias_initializer='zeros'))
    model.add(Activation(activation))    
    if dropout:
        model.add(Dropout(pdrop))
    if batch_normalization:
        model.add(BatchNormalization())  
    
    model.add(Dense(50, kernel_initializer='he_uniform',bias_initializer='zeros'))
    model.add(Activation(activation))    
    if dropout:
        model.add(Dropout(pdrop))
    if batch_normalization:
        model.add(BatchNormalization())      
   
    model.add(Dense(num_classes))
    model.add(Activation('softmax',name='CNN_HumanActionRecognition'))
    
    return model
```

```python
one_image = (64, 128, 3)
num_classes=15
units = 32
activation = 'sigmoid'


def train_test_model(hparams):

    model=architecture(input_shape=one_image, num_classes=num_classes,
                       units=hparams[HP_NUM_UNITS], activation = activation,
                       pdrop=hparams[HP_DROPOUT])
    model.compile(
          optimizer=hparams[HP_OPTIMIZER],
          loss='categorical_crossentropy',
          metrics=['accuracy']
      )
    model.optimizer.learning_rate=hparams[HP_LEARNINGRATE]
    model.fit(x_train, y_train, epochs = 3, batch_size = 32, \
                                validation_split = 0.1, shuffle=True, verbose=0) # Run with 1 epoch to speed things up for demo purposes
    cost, accuracy = model.evaluate(x=x_test, y=y_test)
    return accuracy
```

```python
def training(run_dir, hparams):
    with tf.summary.create_file_writer(run_dir).as_default():
        hp.hparams(hparams)  # record the values used in this trial
        accuracy = train_test_model(hparams)
        tf.summary.scalar(METRIC_ACCURACY, accuracy, step=1)
```

```python
from collections import OrderedDict, namedtuple
from itertools import product
session_num = 0

class RunBuilder():
    @staticmethod
    def get_runs(params):
        Run=namedtuple('Run', params.keys())

        runs=[]

        for v in product(*params.values()):
            runs.append(Run(*v))

        return runs
    
params=OrderedDict(
        num_units=[num_units for num_units in HP_NUM_UNITS.domain.values],
        p_drop=[dropout_rate for dropout_rate in (HP_DROPOUT.domain.values)],
        optim=[optimizer for optimizer in HP_OPTIMIZER.domain.values],
        lr=[lr for lr in HP_LEARNINGRATE.domain.values],
    )

for run in RunBuilder().get_runs(params):
    hparams = {
          HP_NUM_UNITS: run.num_units,
          HP_DROPOUT: run.p_drop,
          HP_OPTIMIZER: run.optim,
          HP_LEARNINGRATE: run.lr
      }
    run_name = "run-%d" % session_num
    print('--- Starting trial: %s' % run_name)
    print({h.name: hparams[h] for h in hparams})
    training('logs/hparam_tuning/' + run_name, hparams)
    session_num += 1
```

```python
one_image = (64, 128, 3)
num_classes=15
units = 32
activation = 'sigmoid'
pdrop = 0.1

best_model = architecture(input_shape=one_image, num_classes=num_classes,
                       units=units, activation = activation,
                       pdrop=pdrop)
```

```python
best_model.summary()
```

```python
learning_rate = 0.01

optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
```

```python
loss_function = 'categorical_crossentropy'
metric_function = 'accuracy'
```

```python
best_model.compile(optimizer = optimizer, loss = loss_function, metrics = [metric_function])
```

```python
#10 % of the training data will be used to validate the training
start_time = time.time()

validation_portion = 0.1
batch_size=64
num_epochs = 40

tensorboard_callback = keras.callbacks.TensorBoard(
    log_dir="logs/best_model", histogram_freq=1)

history = best_model.fit(x=x_train, y=y_train, epochs=num_epochs, batch_size=batch_size, \
                            validation_split=validation_portion, shuffle=True,
                            callbacks=[tensorboard_callback], verbose=1)

end_time = time.time()
print("Time for training: {:10.4f}s".format(end_time - start_time))
```
